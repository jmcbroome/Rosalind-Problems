```python
def kmer_generator(length, seq):
    '''Takes an input string and produces all linear combinations of substrings of length given, e.g. "2, ACG" would output "AC, "CG".'''
    kmer_list = []
    for index in range(0, len(seq)):
        if len(seq[index:index+length]) == length:
            kmer_list.append(seq[index:index+length])
    return kmer_list
        
        
```
